```python
from scipy.stats import ttest_ind

Relation_addiction=Stud_plat[Stud_plat['Relationship_Status']=='In Relationship']['Addicted_Score']
Single_addiction=Stud_plat[Stud_plat['Relationship_Status']=='Single']['Addicted_Score']

t_stat,p_val=ttest_ind(Relation_addiction,Single_addiction,equal_var=False)

n_relation = len(Relation_addiction)
n_single = len(Single_addiction)
print("Sample Size - In Relation:",n_relation)
print("Sample Size - Single:",n_single)

print ("T-statistic:",round(t_stat,3))
print ("P-value:",round(p_val,3))
```

```python
if p_val>0.05:
    print("Failed to reject Null Hypothesis: No significant difference in addiction score between single and in relationship students")

else:
    print("Reject Null Hypothesis: difference in addiction score between single and in relationship students")
```

```python
Female_addiction=Stud_plat[Stud_plat['Gender']=='Female']['Addicted_Score']
Male_addiction=Stud_plat[Stud_plat['Gender']=='Male']['Addicted_Score']

t_stat,p_val=ttest_ind(Female_addiction,Male_addiction,equal_var=False)

n_male = len(Male_addiction)
n_female = len(Female_addiction)
print("Sample Size - Male:",n_male)
print("Sample Size - Female:",n_female)

print ("T-statistic:",round(t_stat,3))
print ("P-value:",round(p_val,3))
```

```python
if p_val>0.05:
    print("Failed to reject Null Hypothesis: No significant difference in addiction score between male and female students")

else:
    print("Reject Null Hypothesis: difference in addiction score between male and female students")
```
